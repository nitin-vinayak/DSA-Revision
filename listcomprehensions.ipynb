## List Comprehensions

```python
squares = [i ** 2 for i in range(5)]
print(squares)
```

```python
numbers = [i for i in range(10)]
print(numbers)
```

```python
fruits = [
    'apple', 
    'banana',
    'pineapple'
]

upper_fruits = [i.upper() for i in fruits]   
print(upper_fruits)
```

```python
scores = [90, 55, 40, 30, 95]

pass_fail = ['Pass' if i >= 40 else 'Fail' for i in scores]
print(pass_fail)
```

## Nested List Comprehensions

```python
matrix = [[1,2,3], [4,5,6]]

flattened = [num for row in matrix for num in row]
print(flattened)
```

```python
table = [[i * j for j in range(5)] for i in range(5)]
print(table)
```

```python
sentence = 'Hello World'    

vowels = []
for char in sentence: 
    if char in 'aeiou':
        vowels.append(char)

print(vowels)

# With Comprehension
vowels_comprehension = [char for char in sentence if char in 'aeiou']
print(vowels_comprehension)
```

```python
text = 'Hello3204'

digits = [digit for digit in text if digit.isdigit()]
print(digits)
```

## Disctionary Comprehensions

```python
squares = {x: x ** 2 for x in range(5)}
print(squares)
```

```python
original = {'a':1, 'b':2, 'c':3}
swapped = {value:key for key,value in original.items()}
print(swapped)
```

## Fahrenheit to Celsius
Convert Fahrenheit to Celsius

```python
temp = [32, 68, 86, 104]

c = [((i-32)*(5/9)) for i in temp]
print(c)
```

## Extract numbers

```python
data = ["abc123", "def456", "ghi789"]

digits = [''.join([digit for digit in s if digit.isdigit()]) for s in data]
print(digits)
```
