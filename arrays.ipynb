## Array Traversal

```python
arr = [1,2,3,4,5,6,7,8,9]   

def forward_traversal(arr): 
    for i in range(len(arr)):
        print(f'{i}: {arr[i]}')

forward_traversal(arr)
```

```python
def backward_traversal(arr): 
    for i in range(len(arr) -1, -1, -1):
        print(f'{i}: {arr[i]}')

backward_traversal(arr)
print('\n')
print('Simpler Version')

# Simpler Version

for i in reversed(range(len(arr))):
    print(f'{i}: {arr[i]}')
```

## Array Insertion

```python
arr = [1,2,3,4,5,6,7,8,9]   

insert_element = 5 
insert_position = 3

print('Before Insertion')   
for i in range(len(arr)):
        print(f'{i}: {arr[i]}')
print('\n')

def insert_into_array(arr, insert_element, insert_position): 
    arr.append(None)
    for i in range(len(arr)-1, insert_position, -1):
        arr[i] = arr[i-1]   
    arr[insert_position] = insert_element
    for i in range(len(arr)):
        print(f'{i}: {arr[i]}')

print('After Insertion')
insert_into_array(arr, insert_element, insert_position)

# Simpler Version

print('\n')
print('Simpler Version')

arr = [1,2,3,4,5,6,7,8,9]   

def insert_at_index(arr, index, value):
    arr.insert(index, value)
    return arr
insert_at_index(arr, insert_position,insert_element)
print(arr)

```

## Array Deletion

```python
arr = [1,2,3,4,5,6,7,8,9]  

index = 5

def delete_from_array(arr, index):
    if 0 <= index <= len(arr):
        val = arr[index]    

    for i in range(index, len(arr)-1):
        arr[i] = arr[i+1]   

    arr.pop()
    return val

print(f'Deleted Value: {delete_from_array(arr, index)}\n')

for i in range(len(arr)):
    print(f'{i}: {arr[i]}')

# Simpler Version

print('\n')
print('Simpler Version\n')

arr = [1,2,3,4,5,6,7,8,9]  

def delete_from_arr(arr, index):
    if 0 <= index < len(arr):
        return arr.pop(index)
    return None

result = delete_from_arr(arr, index)

print(f'Deleted Value: {result}\n')
for i in range(len(arr)):
    print(f'{i}: {arr[i]}')

```

## Linear Search an Array

```python
arr = [1,2,3,4,5,6,7,8,9]

value = 6

def linear_search(arr, value): 
    for i in range(len(arr)):
        if arr[i] == value:
            return i
    return None

print(f'Index Position: {linear_search(arr, value)}')
```

## Binary Search an Array

```python
arr = [1,2,3,4,5,6,7,8,9]

value = 2

def binary_search(arr, value):
    left, right = 0, len(arr)-1 
    while left <= right: 
        
        mid = (left + right) // 2
        
        if arr[mid] == value:
            return mid
        elif arr[mid] < value:
            left = mid + 1
        else: 
            right = mid - 1
            
    return None 

print(f'Index Position {binary_search(arr, value)}') 
```

## Two Pointer Pattern

### Reverse an Array

```python
arr = [1,2,3,4,5,6,7,8,9]

def reverse_arr(arr):
    left, right = 0, len(arr)-1
    if left < right: 
        arr[left] = arr[right]
        left += 1
        right -= 1
    return arr

print(f'Before Reversed: {arr}')
print(f'After Reversed: {reverse_arr(arr)}')
```

### Find the First Instance of a Target Sum in a Sorted Array

```python
def two_sum_sorted(arr, target):
   
    left, right = 0, len(arr) - 1
    
    while left < right:
        current_sum = arr[left] + arr[right]
        
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    
    return []

arr = [1,2,3,3,4,5,6,7,8,9]
print(f'Pair: {two_sum_sorted(arr, 8)}')
```

### Remove Duplicates

```python
arr = [1,1,1,2,2,2,3,4,5,6,7,7,7,7,8,9]

def remove_duplicates (arr):
    if len(arr) == 0:
        return None
    else: 
        l = 1
        for r in range(1, len(arr)): 
            if arr[r] != arr[r-1]:
                arr[l] = arr[r]
                l += 1
        return arr[:l]

print(remove_duplicates(arr))
```

## Sliding Window Pattern


### Max sum of a Sub Array

```python
def max_sum_array(arr, k):
    if len(arr) < k:
        return None
    window_sum = sum(arr[:k])
    max_sum = window_sum
    for i in range(k, len(arr)):
        window_sum = window_sum - arr[i-k] + arr[i]
        max_sum = max(window_sum, max_sum)
    return max_sum

arr = [1,1,1,2,2,2,3,4,5,6,7,7,7,7,8,9]

print(max_sum_array(arr, 2))
```
