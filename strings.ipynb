## String Operations

```python
s = 'Hello' 

print(s[0],'\n')

print(s[:3], '\n')

print(s[2:4], '\n')

print(s[::-1], '\n')

print(s[::2], '\n')
```

## Checking Strings

```python
s = 'Hello World' 


print(s.upper())

print(s.lower())

print(s.isdigit())

print(s.isalpha())

print(s.swapcase())

print(s.startswith('He'))

print(s.endswith('llo'))
print(s.endswith('orld'))
```

## Searching Strings

```python
print(s.find('World')) # Prints the index address of the first occurance

print(s.find('xusdaf'))

print(s.count('l')) # Counts the number of occurances
```

## Splitting and Joining Strings

```python
print(s.split())

print(s.split('o'))

words = s.split()
print('-'.join(words))
```

## Removing Space and characters from Strings

```python
s2 = "  Hello  "

print(s2.strip())
print(s2.lstrip())
print(s2.rstrip())

s3 = "xxxHelloxxx"
print(s3.strip('x'))
print(s3.lstrip('x'))
```

## Replacing

```python
print(s.replace("World", "Python"))
```

## String Building

```python
# Append to list and then join on empty space
chars = []

for i in 'Hello World':
    chars.append(i)
result = ''.join(chars)

print(result)



# Even faster is to use list comprehensions
result_comprehension = ''.join([i for i in 'Hello World'])
print(result_comprehension)
```

## String Manipulation

```python
def reverse_string(s): 
    return s[::-1]
print(reverse_string('Hello'))


def remove_spaces(s):
    return s.replace(" ", "")
print(remove_spaces('Hello World'))


def remove_chars(s, chars):
    return ''.join([c for c in s if c not in chars])
print(remove_chars('Hello World', 'HW'))


def capitalize_first_letter(s): 
    words = s.split()
    return ' '.join(word.capitalize() for word in words)
print(capitalize_first_letter('hello world'))
```
