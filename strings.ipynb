## String Operations

```python
s = 'Hello' 

print(s[0],'\n')

print(s[:3], '\n')

print(s[2:4], '\n')

print(s[::-1], '\n')

print(s[::2], '\n')
```

## Checking Strings

```python
s = 'Hello World' 


print(s.upper())

print(s.lower())

print(s.isdigit())

print(s.isalpha())

print(s.swapcase())

print(s.startswith('He'))

print(s.endswith('llo'))
print(s.endswith('orld'))
```

## Searching Strings

```python
print(s.find('World')) # Prints the index address of the first occurance

print(s.find('xusdaf'))

print(s.count('l')) # Counts the number of occurances
```
