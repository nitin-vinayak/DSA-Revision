## FinBuzz
Print numbers 1-100, but for multiples of 3 print "Fizz", for multiples of 5 print "Buzz", and for multiples of both print "FizzBuzz"

```python
for i in range(100):
    if i == 0:
        print(i)
        continue
    elif i % 3 == 0 and i % 5 == 0:
        print(f'{i}: FinBuzz')
    elif i % 3 == 0:
        print(f'{i}: FIzz')
    elif i % 5 == 0: 
        print(f'{i}: Buzz')
    else: 
        print(i)
```

## Random Number Guesser 
Generate a random number (1-100) and let the user guess until they get it right, giving "higher" or "lower" hints

```python
import random
i = random.randint(0,100)

user_input = int(input("Enter a number"))
print(i)

while i != user_input:
    if i < user_input:
        print('Lesser')
        user_input = int(input("Enter a number"))
    elif i > user_input:
        print('Greater')
        user_input = int(input("Enter a number"))
    else:
        print('You guessed it!!')
```

## Prime Number Checker 
Write a program that checks if a number is prime

```python
number = 11

if number < 2:
    print('Not Prime')
elif number == 2:
    print('Not Prime')
elif number % 2 == 0:
    print('Not Prime')

else:
    is_prime = True
    for i in range(3, int(number**0.5)+1, 2):
        if number % i == 0:
            is_prime = False
            break
    
    if is_prime:
        print('Prime')
    else:
        print('Not Prime')
```
