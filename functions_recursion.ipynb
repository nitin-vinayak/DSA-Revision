## Recursions

```python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    else: 
        return n*factorial(n-1)

factorial(5)
```

```python
numbers = [2,3,4,10]
def sum_numbers(numbers):
    if len(numbers) == 0:
        return 0
    else:
        return numbers[0] + sum_numbers(numbers[1:])
    
sum_numbers(numbers)
```

## Reverse String 
Write a recursive function to reverse a string

```python
string = 'hello'

def reverse_string(string):
    if len(string) == 0:
        return string
    else: 
        return string[-1] + reverse_string(string[:-1])

print(reverse_string(string))
```

## Palindrome Checker
Write a recursive function to check if a string is a palindrome

```python
pal = 'nitin'

def palindrome(pal):
    if len(pal)  <= 1:
        return True
    elif pal[0] == pal[-1]:
        return palindrome(pal[1:-1])
    else: 
        return False
    
palindrome(pal)
```
